```python
top_10_regions = df1[df1['SizeRank'] <= 10]
```

```python
grouped_data = top_10_regions.groupby('RegionName')

```

```python
average_home_values = grouped_data['2024-01-31'].mean()
print(average_home_values)
```

```python
top_10 = average_home_values.sort_values(ascending=False)
top_df = top_10.to_frame().reset_index()
top_df['2024-01-31'] = top_df['2024-01-31'].map('${:,.2f}'.format)
top_df['2024-01-31'] = top_df['2024-01-31'].str.replace('[\$,]', '', regex=True)
top_df['2024-01-31'] = top_df['2024-01-31'].astype(float)
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=top_df, x='RegionName',y='2024-01-31')
plt.title('Average Home Values in Top 10 Most Populous Regions')
plt.xlabel('Region')
plt.ylabel('Average Home Value')
plt.tight_layout()

# for p in ax.patches:
#  ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
ax.bar_label(ax.containers[0], fmt='${:,.2f}')

plt.xticks(rotation=45)


plt.show()
```

```python
price_units_df = df2.groupby('Price per Unit')['Units Sold'].sum().reset_index()


plt.figure(figsize=(10, 6))
plt.scatter(price_units_df['Price per Unit'], price_units_df['Units Sold'])
plt.title('Price vs Units Sold')
plt.xlabel('Price per Unit')
plt.ylabel('Sum of Units Sold')
plt.grid(True)
plt.tight_layout()
plt.show()
```
